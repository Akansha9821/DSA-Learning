```python
# Duplicates Not Allowed
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020
}
print(thisdict)
```

```python
# Accessing Items
# You can access the items of a dictionary by referring to its key name, inside square brackets

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020
}
print(thisdict["brand"])
print(thisdict.get("year"))
x = thisdict.items()
print(x)
```

```python
# Change Values
#  You can change the value of a specific item by referring to its key name:
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

thisdict["year"] = 2021
print(thisdict["year"])

thisdict.update({"model": "suzuki"})
print(thisdict["model"])
```

```python
# Adding Items
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

thisdict["color"] = ["red", "white", "black"]
print(thisdict)

thisdict.update({"price":23})
print(thisdict)
```

```python
# Removing Items
# here are several methods to remove items from a dictionary:
d= {'brand': 'Ford', 'model': 'Mustang', 'year': 1964, 'color': ['red', 'white', 'black'], 'price': 23}

# 1. pop() method removes the item with the specified key name:

d.pop("year")
print(d)


# 2. The popitem() method removes the last inserted item
d.popitem()
print(d)

# 3. The del keyword removes the item with the specified key name:
del d["brand"]
print(d)

# 4. The clear() method empties the dictionary
d.clear()
print(d)
```

```python
# Loop Through a Dictionary
# You can loop through a dictionary by using a for loop.

# When looping through a dictionary, the return value are the keys of the dictionary, but there are methods to return the values as well.

# 1. using for loop
d = {'brand': 'Ford', 'model': 'Mustang', 'year': 1964, 'color': ['red', 'white', 'black'], 'price': 23}
for i in d:
    print("**********using for loop get itrate dict keys:->",i)
    print('&&&&&&&&&&&&&&&&& iterate dict value:->', d[i])
    
    
# 2. using key value function
for k in d.keys():
    print(k)

for v in d.values():
    print(v)
    
    
# 3. using items method
for ke, val in d.items():
    print("key : ->", ke, "value: ->", val)
    
    
```

```python
# Copy a Dictionary
#  You cannot copy a dictionary simply by typing dict2 = dict1, because: dict2 will only be a reference to dict1, and changes made in dict1 will automatically also be made in dict2.

d = {1:"one", 2:"two", 3:"three"}
d1 = d
d1[1] = "four"
print(d)
print(d1)


# 1. There are ways to make a copy, one way is to use the built-in Dictionary method copy().

d2 = d.copy()
d2[3] = "five"
print("+++++++++++++",d2)
print('*********', d)



# 2. Make a copy of a dictionary with the dict() function
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
mydict = dict(thisdict)
print(mydict)
```

```python
# Nested Dictionaries
# A dictionary can contain dictionaries, this is called nested dictionaries.

d1 = {1: 'four', 2: 'two', 3: 'three'}
d2 = {'brand': 'Ford', 'model': 'Mustang', 'year': 1964}
d3 ={'brand': 'Ford', 'model': 'Mustang', 'year': 1964, 'color': ['red', 'white', 'black'], 'price': 23}

nesteddict = {
    "d1": d1,
    "d2": d2,
    "d3":d3
}
print(nesteddict[d1])
```
