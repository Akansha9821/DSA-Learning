```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class LLD:
    def __init__(self):
        self.head = None
        
    def insert_node(self, data):
        n = Node(data)
        n.next = self.head
        self.head = n 
        
    def print_list(self):
        h = self.head
        while h != None:
            print(h.data)
            h = h.next
            
            
    def insert_begin(self, data):
        n = Node(data)
        if n == None:
            self.head = n
        else:
            h = self.head
            while h.next != None:
                h = h.next
            h.next = n
            
            
    def insert_end(self, data):
            self.insert_node(data)
            
            
    def insert_indx(self, data, x):
        n = Node(data)
        h= self.head
        c = 1
        if h ==None:
            self.insert_node(data)
        else:
            while c < x-1:
                h = h.next
                c +=1
                
            n.next = h.next
            h.next= n
            
            
    def delete_begin(self):
        self.head = self.head.next
        
        
    def delete_end(self):
        h = self.head
        while h.next.next != None:
            h = h.next
        h.next = None
        
        
    def delete_index(self, i):
        if self.head == None:
            print("Empty Linked list can't delete node")
        
        if i == self.head.data:
            self.head = self.head.next
            
        else:
            h = self.head
            while h.next != None:
                if i == h.next.data:
                    break
                h = h.next
                
            
                
            
            
            
            
    
       
            
        
        
    
        
```
